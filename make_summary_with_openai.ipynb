```python
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

json_files = glob.glob("data/ai_hub_손해배상자/*.json")
all_results = []


system_prompt_template = '''너는 판결문을 보고 판결문을 요약해주는 어시스턴트야.
판례 내용의 경우 처음에 "주문"이 있을 수 있고, 그 이후에는 "청구취지", "이유"가 나와.
각각의 주제에 대해서는 글머리기호로 구분되어있고, 그 밑에는 내용이 있어
판례 요약은 아래을 참고해서 작성해줘.

참고사항:
  - 상고인, 피상고인
    - 상고인과 피상고인이 원고인지 피고인지 '원고', '피고' 둘 중 하나로만 명시해줘.
    - 인정사실 혹은 기초사실이 있는 경우 손해배상책임을 참고하여 결론을 요약해줘.
  - 요약
    - 금액적인 내용은 포함하지 마.
    - 처음 두 줄은 분석을 위한 핵심내용을 기반으로 summary에 작성해줘.
    - 주제(예시: "기초 사실", "청구에 관한 판단", "원고 주장", "피고 주장", "소결론, "결론")가 있을 경우 주제마다의 내용을 요약해줘

출력 양식:
{{
  "case_num": "{case_num}",
  "상고인": ...,
  "피상고인": ...,
  "비용 청구 내용": 소송비용 중 누구가 얼마나 부담하는지 명시해줘.
  "요약": [
    ...,
    ...,
    ...
}}
  
판결문:
{content}
'''

for json_file in tqdm(json_files[10:20]):
  with open(json_file, "r", encoding="utf-8") as f:
    data = json.load(f)
    sentences = data["sentences"]
    case_num = data['doc_id'].split('-')[1]

    system_prompt = system_prompt_template.format(
      # content="\n".join(sentences),
      content=sentences,
      case_num=case_num
    )

    try:
      response = client.chat.completions.create(
        model="gpt-4o-mini",  # 또는 gpt-3.5-turbo
        messages=[
          {"role": "system", "content": system_prompt},
        ],
        temperature=0,
        max_tokens=10000
      )
    except Exception as e:
      print(f"Error: {e}")
      continue
      
    response_data = response.choices[0].message.content
    if response_data is None:
      continue
    else:
      with open(f"data/open_ai_summary/{case_num}.json", "w", encoding="utf-8") as f:
        f.write(response_data)
```

```python
import pandas as pd
import os
from tqdm.auto import tqdm
import json

DATA_PATH = 'pre_data/totality_data/'
data_list = os.listdir(DATA_PATH)
total_df = pd.DataFrame()
for file in tqdm(data_list):
  with open(f"{DATA_PATH}{file}", "r", encoding='utf-8') as f:
    data = json.load(f)
  if ('손해배상' in data['사건명']) and (',' not in data['사건번호']):
    df = pd.DataFrame(data, index=[0])
    total_df = pd.concat([total_df, df], ignore_index=True)

total_df.to_csv('pre_data/totality_data/totality_data.csv', encoding='utf-8')
```
